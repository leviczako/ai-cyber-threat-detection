# AI Cyber Threat Detection - Data Exploration

This notebook explores a cybersecurity-related dataset and prepares it for a simple machine learning threat detection model.

The goal is to understand the structure of the data, identify the target label, and prepare the dataset for training.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import fetch_kddcup99
```

```python
# Load a cybersecurity-related network intrusion dataset
# KDD Cup 99 is a classic dataset used for network intrusion detection research

data = fetch_kddcup99(percent10=True, as_frame=True)

df = data.frame

df.head()
```

```python
# Check dataset shape

df.shape
```

```python
# Check column names

df.columns
```

```python
# Check basic information about the dataset

df.info()
```

```python
# Check missing values

df.isnull().sum()
```

```python
# Check target label distribution

df["labels"].value_counts().head(20)
```

```python
# Create a simplified binary target:
# normal = 0
# attack/suspicious = 1

df["target"] = df["labels"].apply(lambda x: 0 if x == b"normal." else 1)

df["target"].value_counts()
```

```python
# Visualize normal vs suspicious activity

df["target"].value_counts().plot(kind="bar")

plt.title("Normal vs Suspicious Network Activity")
plt.xlabel("Class")
plt.ylabel("Number of Records")
plt.xticks([0, 1], ["Normal", "Suspicious"], rotation=0)
plt.show()
```

## Initial Findings

The dataset contains network connection records with labels indicating whether the activity is normal or related to an attack.

For the first version of this project, the original labels were simplified into a binary classification problem:

- `0` = normal network activity
- `1` = suspicious or malicious activity

This makes the project easier to understand and suitable for a first AI-based cybersecurity portfolio project.
